# Python

Python es un lenguaje de programación interpretado de código abierto, multiparadigma, aunque principalmente orientado a objetos de alto nivel.

- **Interpretado**: Ejecuta el código directamente, sin necesidad de compilación.


- **Código abierto**: También llamado **"open source"**, se refiere a un lenguaje que se puede distribuir libremente y puede ser usado sin ninguna restricción.


- **Multiparadigma**: Soporta varios paradigmas de programación como: orientación a objetos, estructurada, programación imperativa y programación funcional (aunque principalmente la orentacion a objetos).


- **Alto nivel**: Son aquellos que se encuentran mas cercanos al lenguaje natural (al lenguaje que usamos para comunicarnos) que al lenguaje maquina.

Python fue creado por **Guido van Rossum**, un programador holandés a finales de los 80, y se caracteriza principalmente por ser un lenguaje de programación fácil y sencillo de aprender. Su código es más organizado, evitando el uso de signos de punto y coma al final de cada sentencia y también evita el uso de llaves para especificar bloques de código.

**Ventajas**:
- Lenguaje de alto nivel.
- Puede ser usado para diversos propóstios, entre los más populares son: 
    - Automatizar tareas.
    - Desarrollo Web.
    - Data Science.
    - Web Scraping (extracción de datos).
    - Desarrollo de interfaz gráfica.
- Amplia colección de librerías y frameworks, entre los cuales destacan **TensorFlow** y **NumPy** (librerías) y **Django** y **Flask** (frameworks).
- Portabilidad: Python es compatible con todos los sistemas operativos.
- Comunidad: Python cuenta con una gran comunidad en todas las diferentes areas en la que es utilizado.

**Desventajas**:
- Lentitud.
- Consumo de memoria.
- Desarrollo móvil: Es posible, pero no el más apropiado.

- **Anaconda Navigator** es una interfaz gráfica de usuario (GUI) de escritorio incluida en la distribución Anaconda que permite iniciar aplicaciones y administrar fácilmente paquetes, entornos y canales conda sin usar comandos.

- **Jupyter Notebook** es una interfaz web de código abierto que permite la inclusión de texto, video, audio, imágenes asi como la ejecución de código a través del navegador en múltiples lenguajes.

# Primeros pasos con Jupyter

Para abrir JupyterLab, debemos ejecutar Anaconda Navigator y buscar el programa. De forma alternativa, podemos ejecutar el comando `jupyter lab` desde nuestra terminal o desde Anaconda Prompt. Se abrirá nuestro navegador donde veremos la interfaz de Jupyter.

![Abrir JupyterLab](resources/01_jupyter.png)


Para crear un Notebook en JupyterLab podemos navegar hacia el directorio donde lo queremos utilizando el panel de navegación que aparece a la izquierda.

![Navegación](resources/02_jupyter.png)

Una vez estemos en la ubicación adecuada, podemos seleccionar con qué entorno queremos crear el Notebook.

![Crear](resources/03_jupyter.png)

Podemos renombrar nuestro notebook.

![Renombrar](resources/04_jupyter.png)

El punto grueso donde quedaría la cruz para cerrar una pestaña nos indica que el Notebook tiene cambios sin guardar. Para guardar podemos clicar en el icono del disquete o pulsar CTRL+S (CMD+S para MacOS).

![Guardar](resources/05_jupyter.png)

Los Notebooks están compuestos por celdas en las que podemos escribir código de Python. Para añadir celdas podemos utilizar el botón `+`.

![Crear celdas](resources/06_jupyter.png)

Podemos escribir código en cualquier celda. Luego de escribirlo, podemos ejecutarlo con el botón de play que aparece arriba. Alternativamente, podemos pulsar CTRL+Enter o SHIFT+Enter para ejecuitar la celda que está seleccionada.

También existe el botón de stop que interrumpe la ejecución de un código, el botón de reinicio que libera toda la memoria utiliazda por el Notebook y el botón de forward que ejecuta todas las celdas del notebook en orden.

![Ejecutar código](resources/07_jupyter.png)

Algunas de las celdas pueden contener texto. Podemos seleccionar de qué tipo será nuestra celda con el siguiente desplegable.

![Tipos de celdas](resources/08_jupyter.png)

Podremos siempre escoger entre código, texto raw (texto simple) o Markdown (texto enriquecido, como el de este mismo documento).

![Markdown](resources/09_jupyter.png)
