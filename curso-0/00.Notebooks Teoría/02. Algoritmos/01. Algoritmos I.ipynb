## Pensamiento Algorítmico I

Un **algoritmo** es una secuencia de pasos **finitos** y **ordenados** que permiten **resolver un problema** o realizar una tarea. En la vida cotidiana usamos algoritmos sin darnos cuenta. Ejemplo:

-   **Cómo preparar un sándwich**
    1. Tomar dos rebanadas de pan.
    2. Untar mantequilla en una de ellas.
    3. Colocar jamón y queso.
    4. Cubrir con la otra rebanada.
    5. Servir el sándwich.

Este ejemplo muestra cómo desglosamos una tarea en pasos finitos y ordenados.

### La importancia de los Algorítmos

El objetivo de los algorítmos es la de proporcionar soluciones estructuradas y eficientes a problemas sencillos o complejos.

Se puede dividir este objetivo en:
1. Dividir el problema en pasos más pequeños y más "fáciles" de resolver.
2. Reducir errores en la ejecución.
3. Facilitar la depuración y modificación.
4. Posibilidad de automatizar procesos.
5. Hacer que la solución sea comprensible para otros.

### Introducción al pseudocódigo

El **pseudocódigo** es una manera estructurada de escribir un algoritmo usando lenguaje natural y términos similares a la programación, pero sin depender de la sintaxis específica de un lenguaje de programación.

**Ejemplo**:
```
Inicio
  Leer edad
  Si edad >= 18 Entonces
    Escribir "Eres mayor de edad"
  Si no
    Escribir "Eres menor de edad"
  Fin Si
Fin
```

Elementos clave:

-   **Inicio y Fin** → Delimitan el algoritmo.
-   **Entrada de datos** → `Leer`.
-   **Condiciones** → `Si - Entonces - Si no`.
-   **Salida de datos** → `Escribir`.

#### Interfaz de PSeInt

PSeInt cuenta con una interfaz organizada en secciones que facilitan tanto la creación como la ejecución del pseudocódigo:

-   **Editor de código:** Área principal para redactar y modificar el pseudocódigo.
  
-   **Ejecución normal:** Botón con icono de _play_ en color verde, que ejecuta el programa de manera completa.
  
-   **Ejecución paso a paso:** Botón identificado por dos huellas, ideal para depurar y analizar el flujo del programa en cada etapa.
  
-   **Vista de diagrama de flujo:** Ubicada justo a la derecha del botón de ejecución paso a paso, permite visualizar gráficamente la lógica del pseudocódigo.
